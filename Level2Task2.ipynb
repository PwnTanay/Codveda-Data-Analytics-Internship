```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
df = pd.read_csv("2) Stock Prices Data Set.csv")

df.head()
```

```python
df["date"] = pd.to_datetime(df["date"])
```

```python
apple = df[df["symbol"] == "AAPL"].copy()

apple.head()
```

```python
apple.set_index("date", inplace=True)

apple.head()
```

```python
plt.figure(figsize=(12,6))

plt.plot(apple.index, apple["close"])

plt.title("Apple Stock Closing Price Over Time")
plt.xlabel("Date")
plt.ylabel("Closing Price")

plt.show()
```

```python
apple["MA7"] = apple["close"].rolling(window=7).mean()

plt.figure(figsize=(12,6))

plt.plot(apple["close"], label="Close Price")
plt.plot(apple["MA7"], label="7-Day Moving Average")

plt.title("Apple Stock with 7-Day Moving Average")

plt.legend()

plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

decomposition = seasonal_decompose(
    apple["close"],
    model="additive",
    period=30
)

decomposition.plot()
plt.show()
```

# Conclusion

Time series analysis was performed on Apple stock prices.

- The closing prices were visualized over time.
- A 7-day moving average was calculated to smooth short-term fluctuations.
- Seasonal decomposition was used to separate the data into trend, seasonal, and residual components.

This analysis helps in understanding long-term trends and recurring patterns in stock prices.
