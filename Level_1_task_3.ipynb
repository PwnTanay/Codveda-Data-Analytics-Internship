```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/1) iris.csv")
```

```python
df.head()
```

BAR PLOT

```python
plt.figure(figsize=(6,4))
df["species"].value_counts().plot(kind="bar")
plt.title("Number of Flowers in Each Species")
plt.xlabel("Species")
plt.ylabel("Count")
plt.savefig("barplot.png")
plt.show()
```

Scatter Plot

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    data=df,
    x="sepal_length",
    y="petal_length",
    hue="species"
)
plt.title("Sepal Length vs Petal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.savefig("scatterplot.png")
plt.show()
```

Line Chart

```python
  plt.figure(figsize=(8,5))

plt.plot(df.index, df["sepal_length"])

plt.title("Sepal Length Trend")
plt.xlabel("Flower Index")
plt.ylabel("Sepal Length")

plt.savefig("linechart.png")
plt.show()
```


Three visualizations were created using the Iris dataset:

1. Bar Plot showing species distribution.
2. Scatter Plot showing the relationship between sepal length and petal length.
3. Line Chart showing sepal length variation across observations.

All plots were customized with titles and labels and exported as image files.
